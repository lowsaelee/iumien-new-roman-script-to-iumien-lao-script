## Convert from Iu Mien new Roman script to unofficial Iu Mien Lao script

```python
# Importing libraries

import codecs
import nltk
import re
import os
```

### Function to read data from text and put into Python list

```python
def add_to_mien_laomien_db(filename, mien, laomien):
    with codecs.open(filename, 'r', encoding='utf8') as f:
        text = f.read()
        
    sentences = re.split('\n|\t', text)
    new_sentences = []
    
    #print(sentences)
    
    openBracket = 0
    for i in range(len(sentences)):
        # stops reading the text file when sentence is equal to this sentence
        # text after this sentence is irrevalent
        if sentences[i] == "Society. ":
            break

        # actual sentences for the corpus begins after the first two brackets which are at the beginning of the sentence
        # counts the openbrackets
        if len(sentences[i]) > 1:
            if sentences[i][0] == "[":
                openBracket += 1
        
        # start appending to sentences to new_sentences right after the two brackets
        if (len(sentences[i]) > 1) and (openBracket == 2):
            new_sentences.append(sentences[i])

    
    verseCount = 0
    for i in range(len(new_sentences)-1):
        if(new_sentences[i][0] == new_sentences[i+1][0] and verseCount <= 7):
            verseCount = verseCount + 1
            #print(verseCount)
            #print(new_sentences[i][0])
            mien.append(nltk.word_tokenize(new_sentences[i]))
            laomien.append(nltk.word_tokenize(new_sentences[i+1]))
        
        if(new_sentences[i][0:2] == new_sentences[i+1][0:2]):
            verseCount = verseCount + 1
            #print(verseCount)
            mien.append(nltk.word_tokenize(new_sentences[i]))
            laomien.append(nltk.word_tokenize(new_sentences[i+1]))
```

### Function to query list with user input

```python
def token_search(token, mien, laomien, inputLang):
    transliteration = []
    #print('Mien db length = ',len(mien))
    #print('LaoMien db length = ',len(laomien))
    if inputLang == "laomien":
        for i in range(len(laomien)):
            for j in range(len(laomien[i])):
                if laomien[i][j].lower() == token.lower():
                    try:
                        #print('found it')
                        #print(laomien[i][j])
                        transliteration.append(mien[i][j])
                    except:
                        pass
                        #print('index for laomien shorter than mien')
        #print(transliteration)
    if inputLang == "mien":
        for i in range(len(mien)):
            for j in range(len(mien[i])):
                if mien[i][j].lower() == token.lower():
                    try:
                        #print('found it')
                        #print(laomien[i][j])
                        transliteration.append(laomien[i][j])
                    except:
                        pass
                        #print('index for laomien shorter than mien')
        #print(transliteration)    
    print('total found',len(transliteration))
    try:
        mostLikelyTerm = max(set(transliteration), key = transliteration.count)
    except:
        mostLikelyTerm = "None"
    
    return mostLikelyTerm
```

### For loop to read text files from directory

```python
dataPath = "mien-laomien/"
dataFile = []
dataCounter = 0
for i in os.listdir(dataPath):
    if ".txt" in i:
        fileName = dataPath + i
        dataFile.append(fileName)
        dataCounter += 1
```

### For loop to call function to read text file into Python list

```python
mien = []
laomien = []
for i in range(len(dataFile)):
    print(dataFile[i])
    add_to_mien_laomien_db(dataFile[i], mien, laomien)
```

### sentence to convert from Iu Mien new Roman script to Iu Mien Lao script

```python
sampleSentence = "meih caux ninh nzaeng jaax zoux haaix nyungc"
```

### Tokenize the sample sentence

```python
sampleList = nltk.word_tokenize(sampleSentence)
```

### For loop to query the tokenized words

```python
convertedLaoMien = []
for i in range(len(sampleList)):
    #try:
    search_word = sampleList[i]
    lookupResult = token_search(search_word, mien, laomien,inputLang = "mien")
    convertedLaoMien.append(lookupResult)
    print("Mien - ", search_word)
    print("Lao Mien - ", lookupResult)
```

### Output query into a sentence

```python
laomien_sentence = ''
for i in range(len(convertedLaoMien)):
    laomien_sentence += convertedLaoMien[i] + ' '

print(sampleSentence)
print(laomien_sentence)
```
